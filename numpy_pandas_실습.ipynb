## Numpy 실습

```python
import numpy as np

A=np.array([1,2,3])
B=np.array([4,5,6])
print("A==",A,"B==",B)
print("vector A shape==",A.shape,"vector B shape==",B.shape)
print("vector A dimension ==",A.ndim,"vector B dimension ==",B.ndim)
```

```python
import numpy as np
A=np.array([1,2,3])
B=np.array([4,5,6])
print("A+B ==",A+B)
print("A-B ==",A-B)
print("A*B ==",A*B)
print("A/B ==",A/B)
```

```python
C=np.array([1,2,3])
print("C.shape ==",C.shape)

C_new=C.reshape(1,3) #1행 3열로 재편성
print("C_new ==",C_new) # [[x y z]]
print("C.shape ==",C_new.shape)
```

```python
D=np.array([[1,2,3],[4,5,6]])
print("D.shape ==",D.shape)
D_new=D.reshape(2,3)
print("D_new ==",D_new)
print("D_new.shape ==",D_new.shape)
```

```python
A=np.array([[1,2,3,4],
            [10,20,30,40],
            [100,200,300,400]])
print(A.shape)
```

```python
# reshape(-1,zzz)

B=A.reshape(-1,3) #-1을 넣는 순간 데이터의 개수와 3이라는 열의 계수에 맞게 배치
print("B.shape ==",B.shape)
print(B)
```

```python
A=np.array([[1,2],
            [3,4],
            [5,6]])
B=A.T
print("A.shape ==",A.shape)
print("B.shape ==",B.shape)
print("A ==",A)
print("A transpose ==",B)
```

```python
# 행렬곱-dot product
A=np.array([[1,2,3],
            [4,5,6]])
B=np.array([[-1,-2],
            [-3,-4],
            [-5,-6]])
C=np.dot(A,B)
print("A.shape =",A.shape,"B.shape =",B.shape)
print("C.shape =",C.shape)
print("C =",C)
```

```python
A=np.array([[10,20],[30,40],[50,60]])

print("A.shape ==",A.shape)
print(A)

print("A[0,0] ==",A[0][0])
print("A[2][1] ==",A[2,1])
print(A[:,0])
print(A[0:-1,1:2])
print(A[:,:])
```

```python
A=np.array([[10,20,30,40],
            [50,60,70,80]])
it=np.nditer(A,flags=['multi_index'],op_flags=['readwrite'])

while not it.finished:
    idx=it.multi_index
    print('index =>',idx,', value =>',A[idx])
    it.iternext()
```

#### numpy 이터레이터를 사용하는 이유에 대해서는 지금 당장 와닿지 않을 수도 있다. 이는 다차원 배열을 효율적으로 순회하기 위해서 사용하는 방법이다. 또 op_flags=[ 'readwrite' ] 라는 문구를 통해 array 내의 값을 수정할 수 있다.

```python
import numpy as np

number1=np.random.rand(3)
number2=np.random.rand(1,3)
number3=np.random.rand(3,1)
print("random_number1 ==",number1)
print("random_number2 ==",number2)
print("random_number3 ==",number3)
```

```python
# argmax를 사용하면 최댓값을 가지는 인덱스를 찾아주고, 이는 차후에 딥러닝 구조에서
# 원핫 인코딩 방식을 적용할때 필요한 함수
A=np.array([2,6,3,1,5,8])
print(np.max(A))
print(np.argmax(A))
```
